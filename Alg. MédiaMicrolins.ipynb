```python
import pandas as pd
import time as t
data_atual = t.localtime()
ano, mes, dia, hora, minuto, segundo, diadasemana, diadoano, isdst = data_atual



class selec_cor:
    OK = '\033[92m' #VERDE
    WARNING = '\033[93m' #AMARELO
    FAIL = '\033[91m' #VERMELHO
    RESET = '\033[0m' #RESETAR CORES
    REVERSE = "\033[;7m" #COR REVERSA
    DETAIL = "\033[1;34m" #AZUL
    BOLD = "\033[;1m" #NEGRITO



def transformar_dia(diadasemana):
    '''Esta função é para transformar o dia que seria trazido como número, para um dia da semana.
    Está relacionado com o dicionário localizado abaixo o qual indica os dias correspondentes'''
    for l, each in enumerate(dicionario_dias) :
        if each == diadasemana:
            resultado = dicionario_dias[l]
            return resultado
        else:
            pass


def adaptar_item(item):
    '''Adapta o item que for menor que 10 para pôr um 0 na frente '''
    if item < 10:
        item = '0' + str(item)
        return item
    else:
        return item
  

def verificar_aprovacao():
    '''Verifica a aprovação do aluno; podendo ter 3 opções de aprovação. 
    Uma, caso passe com a média maior ou igual à 95,
     outra caso passe mas seja menor qque 95 e outra caso não passe'''
    if  media_final >= meta_media:
        resul = f'|{nome_aluno:^30}|{curso:^25}|{qnt_aula:^15}|' + f'\033[42m{media_final:^10.2f}\033[0m|'.replace('.',',')

        if media_final >= reqde_certify:
            resul = f'|{nome_aluno:^30}|{curso:^25}|{qnt_aula:^15}|' + f'\033[44m{media_final:^10.2f}\033[0m|'.replace('.',',')

    else:
        resul = f'|{nome_aluno:^30}|{curso:^25}|{qnt_aula:^15}|' + f'\033[41m{media_final:^10.2f}\033[0m|'.replace('.',',')

    return resul



dicionario_dias = { 
0: 'Segunda-feira',
1: 'Terça-feira',
2: 'Quarta-feira',
3: 'Quinta-feira',
4: 'Sexta-feira',
5: 'Sábado',
6: 'Domingo',
}

hoje = f'{adaptar_item(dia)}/{adaptar_item(mes)}/{ano}'
reqde_certify = 95
meta_media = int(70)
nome_validator = ['Lucas', 'Matheus']
cpf_validator = ['123', '000']


while True:
    nome_procurador = input('Insira seu nome: ').strip().title()
    if nome_procurador in nome_validator:
        pass
    else:
        print('Acesso negado. Seu nome não está no nosso banco de dados.')
        break

    #cpf do procurador
    senha_procurador = input("Digite a sua senha: ").strip()
    
    #validador de login e CPF
    indice_login = nome_validator.index(nome_procurador)
    indice_senha = cpf_validator.index(senha_procurador)


    if indice_login == indice_senha:
        pass
    else:
        print('Indexação Divergente')
        break

    #nome do aluno
    nome_aluno = str(input('Informe o nome do aluno: \n')).strip().title()
    if nome_aluno == '':
        print( selec_cor.FAIL + 'Error N!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')
        break
    else:
        pass

    #curso que o aluno está fazendo
    curso = str(input('Qual o curso do aluno {} está querendo saber a média? \n'
    .format(nome_aluno))).strip().title()

    if curso == '':
        print( selec_cor.FAIL + 'Error C!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')
        break
    
    elif len(curso) <= 3:
        curso = curso.upper()

    else:
        pass

    #quantidade de aulas do curso inserido anteriormente 
    qnt_aula = int(input('Insira a quantidade de aulas que o curso de {} tem: \n'.format(curso)).strip())
    if qnt_aula <= 36: #36 é a maior quantidade de aula de um único curso para onde formulei este algoritmo 
        pass

    else:
        print( selec_cor.FAIL + 'Error Q!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')
        break

    if nome_aluno and curso and qnt_aula:
    
        #inserir notas
        contadorwhile = 0
        soma = 0
        while contadorwhile < int(qnt_aula):

            soma += float(input('Insira a nota que o aluno(a) {a} tirou correspondente à AULA {c}.\n(Total de aulas restantes a serem preenchidas: {q})\n'
            .format(q = int(qnt_aula) - contadorwhile, c = contadorwhile + 1, a = nome_aluno)))
            contadorwhile += 1

        media_final = soma / int(qnt_aula)

        print('|{:^30}|{:^25}|{:^15}|{:^10}|'.format('Nome do Aluno', 'Curso', 'Qtde Aula','Média'))
        print('=' * 85)
        print(verificar_aprovacao())
        print('=' * 85)
        
        texto = f"Pesquisa realizada em {hoje} às {adaptar_item(hora)}h{adaptar_item(minuto)}min."
        print('\033[44m{:^85}\033[0m'.format(texto))
    break
```

```python
nomes_alunos = []
info_curso = []


nomes_alunos.append(nome_aluno)
info_curso.append((curso, media_final, hoje))


lista_alunos = dict(zip(nomes_alunos, info_curso))

lista_alunos_df = pd.DataFrame.from_dict(lista_alunos, orient='index')
lista_alunos_df.columns = ["Curso", "Média", 'Data']
lista_alunos_df['Data'] = pd.to_datetime(lista_alunos_df['Data'], format='%d/%m/%Y')
lista_alunos_df.info()
lista_alunos_df.to_csv('Teste do Lucas.csv', sep = ';', encoding='latin1')

```
